```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Ma'lumotlarni yuklash
iris = load_iris()

# Faqat petal uzunligi va kengligini olish
iris_petal = iris.data[:, [2, 3]]
labels = iris.target
target_names = iris.target_names

# Grafik chizish
plt.figure(figsize=(8, 6))
colors = ['red', 'green', 'blue']

for i in range(3):  # 3 ta tur (setosa, versicolor, virginica)
    plt.scatter(
        iris_petal[labels == i, 0],  # petal length
        iris_petal[labels == i, 1],  # petal width
        label=target_names[i],
        color=colors[i]
    )

plt.xlabel('Petal length (cm)')
plt.ylabel('Petal width (cm)')
plt.title('Iris gullari: Petal uzunligi va kengligi')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Iris dataset
iris = load_iris()
iris_petal = iris.data[:, [2, 3]]  # Petal length va width
true_labels = iris.target

# KMeans modelini yaratish (3 cluster – chunki 3 tur bor)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(iris_petal)  # Modelni o‘qitish

# Natijalar
predicted_labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Grafik
plt.figure(figsize=(8, 6))
colors = ['red', 'green', 'blue']

for i in range(3):
    plt.scatter(
        iris_petal[predicted_labels == i, 0],
        iris_petal[predicted_labels == i, 1],
        label=f'Cluster {i+1}',
        color=colors[i],
        alpha=0.6
    )

# Klaster markazlarini ko‘rsatish
plt.scatter(
    centers[:, 0], centers[:, 1],
    s=200, c='black', marker='X',
    label='Cluster Centers'
)

plt.xlabel('Petal length (cm)')
plt.ylabel('Petal width (cm)')
plt.title('K-Means bilan Iris klasterlash (Petal bo‘yicha)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Ma'lumot
iris = load_iris()
iris_petal = iris.data[:, [2, 3]]

# Inertia qiymatlari (K=1 dan 10 gacha)
inertia = []
K = range(1, 11)

for k in K:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(iris_petal)
    inertia.append(model.inertia_)

# Grafik chizish
plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'bo-', linewidth=2)
plt.xlabel('Klaster soni (k)')
plt.ylabel('Inertia (yopishish darajasi)')
plt.title('Elbow method - Optimal K aniqlash')
plt.grid(True)
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np

# Ma'lumot
iris = load_iris()
iris_petal = iris.data[:, [2, 3]]

# Inertia qiymatlari
inertia = []
K = range(1, 11)

for k in K:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(iris_petal)
    inertia.append(model.inertia_)

# Grafik chizish
plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'bo-', linewidth=2)
plt.xlabel('Klaster soni (k)')
plt.ylabel('Inertia')
plt.title('Elbow method - Optimal klaster soni')
plt.grid(True)
plt.show()

# Tizza (elbow) nuqtasini taxminan topish uchun farqlarni hisoblaymiz
diffs = np.diff(inertia, 2)  # 2-chi tartibli farq (egri chiziq)
elbow_point = np.argmax(diffs) + 2  # +2 chunki diff ikkinchi farq, index 0 dan

print(f"👉 Optimal klaster soni (Elbow method bo‘yicha): {elbow_point}")
```

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Ma'lumotlarni yuklash
iris = load_iris()
iris_data = iris.data
iris_petal = iris_data[:, [2, 3]]  # Petal uzunligi va kengligi

# PCA bilan ma'lumotlarni 2 o'lchovga kamaytirish
pca = PCA(n_components=2)
iris_petal_pca = pca.fit_transform(iris_petal)

# KMeans klasterlash (3 klaster)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(iris_petal_pca)
predicted_labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Grafik chizish
plt.figure(figsize=(8, 6))

# KMeans klasterlarini vizualizatsiya qilish
colors = ['red', 'green', 'blue']
for i in range(3):
    plt.scatter(
        iris_petal_pca[predicted_labels == i, 0],
        iris_petal_pca[predicted_labels == i, 1],
        label=f'Cluster {i+1}',
        color=colors[i],
        alpha=0.6
    )

# Klaster markazlarini ko‘rsatish
plt.scatter(
    centers[:, 0], centers[:, 1],
    s=200, c='black', marker='X',
    label='Cluster Centers'
)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA + KMeans - Iris Klasterlash')
plt.legend()
plt.grid(True)
plt.show()
```
